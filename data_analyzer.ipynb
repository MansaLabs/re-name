#### Read names from csv file

```python
def csv_reader(file_obj):
    """
    Read a csv file
    """
    reader = csv.reader(file_obj)
    all_rows = []
    for i,row in enumerate(reader):
        if i == 0:
            continue
        all_rows.append((" ".join(row)).lower())
    return all_rows
```

```python
def do_csv_file(file_path):
    with open(file_path, encoding='utf8') as file:
        rows_list = csv_reader(file)
    all_words_csv = []
    for i, lines in enumerate(rows_list):
        line_words = lines.split(' ')
        for word in line_words:
            if len(word)>2:
                all_words_csv.append(word)
    return all_words_csv
```

```python
def read_txt_file(file_path):
    """
    read the contents of a text file line by line into a list of lines
    each line is also a list of words
    file_path: the location of the text file
    returns: the lines list
    """
    with open(file_path, encoding='utf8') as file:
        text = file.readlines()
        lines_list = [line.strip() for line in text]
    return lines_list
```

```python
def do_txt_file(file_path):
    name_list = read_txt_file(file_path)
    all_words = []
    for i, lines in enumerate(name_list):
        if len(lines)>=1: # skip empty lines
            if lines[0]=='#': # skip commented lines
                continue
            all_words.append(lines.lower())
    return all_words
```

#### Remove duplicates

```python
def remove_duplicates(dup_list):
    new_list = []
    for i in dup_list:
      if i not in new_list:
        new_list.append(i)
    return new_list
```

```python
# list files in the directory
dir_path = './dataset/'
text_files = os.listdir(dir_path)
# read each file and extract names
all_names = []
for files in text_files:
    print(dir_path+files)
    if files[-3:]=='txt':
        
        names_list = do_txt_file(dir_path+files)
    else:
        
        names_list = do_csv_file(dir_path+files)
    # combine names into one list
    # remove duplicates
    all_names +=names_list

# perform no duplicates on all names    
all_names_no_dup = remove_duplicates(all_names)
```

```python
def isAscii(s):
    """
    Remove non ascii characters
    """
    try:
        s.encode('ascii')
    except UnicodeEncodeError:
        return False
    else:
        return True

# examples
print(isAscii('सानिया'))
print(isAscii('my_name'))
```

```python
def contain_vowel(s):
    """
    Check if there is a vowel in a string
    """
    vowels = 'aeiou'
    for letters in vowels:
        if letters in s:
            return True
    return False
```

```python
trimmed_names = []
for names in punct_names:
    newws = deepcopy(names)
    for symb in punct_n_digits:
        newws = newws.replace(symb, '')
    names_splitted = newws.split('-')
    for names in names_splitted:
        if len(names)>2 or contain_vowel(names):
            if len(names)>1:
                trimmed_names.append(names)
```
